# List Comprehension

```python
df.columns = [s.replace('.','_')for s in df.columns]
```

```python
df['variety'].value_counts()
```

## Observação: estão bem distribuídas.
## E as medidas?

```python
df_setosa = df.query('variety == "Setosa"').copy()
df_versicolor = df.query('variety == "Versicolor"').copy()
df_virginica = df.query('variety == "Virginica"').copy()
```

```python
medidas = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
```

```python
print('Setosa')
df_setosa[medidas].hist()
```

```python
print('Versicolor')
df_versicolor[medidas].hist()
```

```python
print('Virginica')
df_virginica[medidas].hist()
```

```python
df.query('petal_length > 5')['variety'].value_counts()
```

# Conclusão: praticamente só as plantas da espécie Virginica têm comprimento de pétala maior que 5!

# Nenhuma amostra da Setosa possui medida maior que 5.

- Isso nos diz que a variável "petal_length" possui um bom poder preditivo referente a "variety".
- Quais amostras são de uma espécie e quais não são.
